# Naive Baiyes
Naive Bayes adalah metode klasifikasi data yang menggunakan prinsip probabilistik (peluang) untuk memprediksi kategori suatu data berdasarkan fitur-fitur yang dimilikinya.

Kenapa disebut "naive" (polos)?
Karena metode ini berasumsi bahwa semua fitur data tidak saling berhubungan (independen). Contohnya, jika kita mengklasifikasi buah berdasarkan warna, bentuk, dan rasa, Naive Bayes akan menganggap warna tidak memengaruhi bentuk atau rasa, meskipun kenyataannya mungkin ada hubungan (misalnya, buah merah biasanya bulat). Asumsi ini membuat perhitungannya jauh lebih sederhana dan cepat.

Contoh Penerapan:
Misalnya, untuk membedakan email "spam" dan "bukan spam":

Fitur yang digunakan bisa berupa kata-kata dalam email (misalnya "hadiah", "gratis").

Naive Bayes menghitung peluang email mengandung kata-kata tersebut dan menentukan apakah itu spam, tanpa memperhatikan hubungan antar kata (misalnya, "hadiah" dan "gratis" dianggap tidak terkait).

Kelebihan:

Cepat dan efisien, cocok untuk data besar atau data sederhana.

Bisa bekerja dengan baik meskipun asumsi "independen" tidak sepenuhnya benar.

#### **2. Rumus Naïve Bayes**
Naïve Bayes menggunakan rumus:

$$
P(C|X) = \frac{P(X|C) P(C)}{P(X)}
$$

Dimana:

( P(C|X) \) adalah probabilitas kelas \( C \) diberikan fitur \( X \)
( P(X|C) \) adalah probabilitas fitur \( X \) muncul dalam kelas \( C \)
( P(C) \) adalah probabilitas prior dari kelas \( C \)
( P(X) \) adalah probabilitas total fitur \( X \)

Karena kita membandingkan antar kelas, kita hanya fokus pada **pembilang** dari rumus tersebut:

$$
P(X|C) P(C)
$$

---
#### **3. Menghitung Mean dan Variance untuk Setiap Kelas**
Karena kita mengasumsikan distribusi normal:

$$
P(X|C) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(X - \mu)^2}{2 \sigma^2}}
$$

Kita hitung **mean (\( \mu \))** dan **variance (\( \sigma^2 \))** dari setiap kelas.

##### **Tabel Mean dan Variance untuk Setiap Kelas**
| Kelas             | Mean Petal Length (\( \mu \)) | Variance Petal Length (\( \sigma^2 \)) | Mean Petal Width (\( \mu \)) | Variance Petal Width (\( \sigma^2 \)) |

| **Iris-setosa**   | 1.46                        | 0.03                           | 0.23                        | 0.01                           |
| **Iris-versicolor** | 4.26                        | 0.22                           | 1.32                        | 0.04                           |
| **Iris-virginica**  | 5.72                        | 0.31                           | 2.03                        | 0.09                           |

---
#### **4. Menghitung Probabilitas untuk Data Baru**
Dengan menggunakan distribusi normal:

$$
P(X|C) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(X - \mu)^2}{2 \sigma^2}}
$$

**Probabilitas Petal Length dan Petal Width untuk setiap kelas:**

| Kelas             | P(PL=4.7 / C) | P(PW=1.8 / C) | P(X / C) |
|------------------|-------------|-------------|------------|
| **Iris-setosa**   | Sangat kecil | Sangat kecil | Sangat kecil |
| **Iris-versicolor** | 0.85        | 0.92        | 0.782 |
| **Iris-virginica**  | 0.45        | 0.89        | 0.401 |

---
#### **5. Menghitung Prior Probabilitas**
Prior probabilitas setiap kelas dihitung dengan:

$$
P(C) = \frac{N_C}{N_{total}}
$$

| Kelas             | Jumlah Data | Prior Probability (\( P(C) \)) |
|------------------|------------|-----------------------|
| **Iris-setosa**   | 50         | 0.33                  |
| **Iris-versicolor** | 50         | 0.33                  |
| **Iris-virginica**  | 50         | 0.33                  |

---
#### **6. Menentukan Kelas dengan Probabilitas Tertinggi**
Akhirnya, kita hitung probabilitas akhir:

$$
P(XC) P(C)
$$

| Kelas             | P(X / C) | P(C) | P(X / C) * P(C) |
|------------------|------------|--------|-----------------|
| **Iris-setosa**   | Sangat kecil | 0.33   | Sangat kecil  |
| **Iris-versicolor** | 0.782       | 0.33   | 0.258        |
| **Iris-virginica**  | 0.401       | 0.33   | 0.133        |

Karena **P(X | Iris-versicolor) * P(Iris-versicolor) adalah yang terbesar**, maka:

$$
\text{Data (4.7, 1.8) diklasifikasikan sebagai: **Iris-versicolor**}
$$

---
#### **Kesimpulan**
Dari perhitungan di atas, kelas dengan probabilitas tertinggi adalah **Iris-versicolor**, sehingga data baru diklasifikasikan ke dalam kelas ini.

#### 1. Menghitung Probabilitas Prior

\[ P(Class) = \frac{Jumlah\ Data\ dalam\ Kelas}{Total\ Data} \]

---

#### 2. Menghitung Mean dan Standard Deviasi untuk Tiap Kelas
Formula:
$$ \mu = \frac{1}{N} \sum x_i $$
$$ \sigma^2 = \frac{1}{N} \sum (x_i - \mu)^2 $$

**a) Mean dan Standar Deviasi untuk Petal Length & Petal Width:**

| Kelas           | Mean Petal Length (\(\mu\)) | Std Dev Petal Length (\(\sigma\)) | Mean Petal Width (\(\mu\)) | Std Dev Petal Width (\(\sigma\)) |
|---------------|----------------------|----------------------|-----------------|------------------|
| Iris-setosa  | 1.462                 | 0.1737               | 0.246          | 0.1054           |

| Iris-virginica | 5.55                 | 0.5519               | 2.026          | 0.2747           |

---

#### 3. Menghitung Probabilitas dengan Distribusi Gaussian
Formula distribusi Gaussian:
$$
P(x|class) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

**Untuk Petal Length = 4.7**

$$ P(4.7 | Setosa) $$ = $$ \frac{1}{\sqrt{2\pi (0.1737^2)}} e^{-\frac{(4.7 1.462)^2}{2(0.1737^2)}} $$ ≈ 0
$$ P(4.7 | Versicolor) $$ = $$ \frac{1}{\sqrt{2\pi (0.4699^2)}} e^{-\frac{(4.7 - 4.26)^2}{2(0.4699^2)}} $$ ≈ 0.6545
$$ P(4.7 | Virginica) $$ = $$ \frac{1}{\sqrt{2\pi (0.5519^2)}} e^{-\frac{(4.7 - 5.55)^2}{2(0.5519^2)}} $$ ≈ 0.3913

**Untuk Petal Width = 1.8**

 $$ P(1.8 | Setosa) $$ = $$ \frac{1}{\sqrt{2\pi (0.1054^2)}} e^{-\frac{(1.8 - 0.246)^2}{2(0.1054^2)}} $$ ≈ 0
 $$ P(1.8 | Versicolor) $$ = $$ \frac{1}{\sqrt{2\pi (0.1978^2)}} e^{-\frac{(1.8 - 1.325)^2}{2(0.1978^2)}} $$ ≈ 0.1937
$$ P(1.8 | Virginica) $$ = $$ \frac{1}{\sqrt{2\pi (0.2747^2)}} e^{-\frac{(1.8 - 2.026)^2}{2(0.2747^2)}} $$ ≈ 1.306

---

#### 4. Menghitung Probabilitas Akhir untuk Setiap Kelas
Formula:
$$ P(Class | Data) \propto P(Class) \times P(x_1 | Class) \times P(x_2 | Class) $$

| Kelas          | Probabilitas Akhir |
|---------------|------------------|
| Setosa       | 0 (dihilangkan karena probabilitas 0) |
| Versicolor   | $$ 0.3333 \times 0.6545 \times 0.1937 $$ ≈ 0.0423 |
| Virginica    | $$ 0.3333 \times 0.3913 \times 1.306 $$ ≈ 0.1703 |

---

#### 5. Kesimpulan
Karena $$ P(Virginica | Data) > P(Versicolor | Data) $$, maka data diklasifikasikan sebagai:

```python
import pandas as pd
from sqlalchemy import create_engine
from scipy.stats import norm  # Perbaikan: Impor norm dari scipy.stats

# Koneksi ke MySQL
mysql_engine = create_engine("mysql+pymysql://avnadmin:AVNS_6Z7yDEzrrMmRqOX1L1N@mysql-37408e5b-irishmysql.h.aivencloud.com:11680/defaultdb")

# Mengambil data dari database
query = "SELECT * FROM iris_mysql;"
df = pd.read_sql(query, mysql_engine)

# Menyesuaikan nama kolom
df.columns = ["id", "class", "petal_length", "petal_width"]

# Data input untuk klasifikasi
input_petal_length = 4.7
input_petal_width = 1.8

# Menghitung probabilitas prior (P(Class))
class_counts = df["class"].value_counts()
total_samples = len(df)
prior_probs = class_counts / total_samples

# Menghitung mean dan standar deviasi untuk setiap kelas
stats = df.groupby("class").agg({
    "petal_length": ["mean", "std"],
    "petal_width": ["mean", "std"]
})

# Rename kolom hasil aggregasi agar lebih mudah diakses
stats.columns = ["petal_length_mean", "petal_length_std", "petal_width_mean", "petal_width_std"]

# Fungsi untuk menghitung probabilitas berdasarkan distribusi Gaussian
def gaussian_prob(x, mean, std):
    return norm.pdf(x, mean, std) if std > 0 else 0  # Hindari std = 0

# Menghitung likelihood untuk setiap kelas
likelihoods = {}
for species in stats.index:
    p_length = gaussian_prob(input_petal_length, stats.loc[species, "petal_length_mean"], stats.loc[species, "petal_length_std"])
    p_width = gaussian_prob(input_petal_width, stats.loc[species, "petal_width_mean"], stats.loc[species, "petal_width_std"])
    likelihoods[species] = prior_probs[species] * p_length * p_width  # P(Class) * P(X|Class)

# Menentukan kelas dengan probabilitas tertinggi
predicted_species = max(likelihoods, key=likelihoods.get)

# Output hasil klasifikasi
print(f"Data ({input_petal_length}, {input_petal_width}) diklasifikasikan sebagai: {predicted_species}")
```
