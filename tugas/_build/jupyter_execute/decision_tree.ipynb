# ðŸŒ³ Penjelasan tentang Decision Tree & Information Gain

## Apa itu Decision Tree?

Bayangin kamu lagi main **20 pertanyaan** buat nebak sesuatu, misalnya: "Apakah hewan ini kucing?". Kamu nanya hal-hal kayak:

- Apakah hewan ini berkaki 4?
- Apakah dia suka mengeong?
- Apakah dia punya bulu?

Nah, **Decision Tree (Pohon Keputusan)** itu mirip banget!  
Dia adalah model yang **membuat keputusan secara bertahap**, mulai dari pertanyaan yang paling "penting", terus bercabang sampai nemu jawaban (kelas).

Strukturnya kayak pohon:
- Akar (root): fitur paling penting yang dipakai buat mulai nanya.
- Cabang (branch): pilihan dari hasil pertanyaan.
- Daun (leaf): hasil akhir â†’ klasifikasi (misalnya: A, B, C).

---

## Terus, apa itu Information Gain?

Waktu kita milih fitur mana yang dipakai di akar pohon, kita pengen milih yang **paling informatif**.  
Nah, itulah tugas si **Information Gain**.

### ðŸ“¦ Gampangnya:
- Information Gain = Seberapa banyak "ketidakpastian" (entropi) yang **berkurang** setelah kita membagi data berdasarkan fitur tertentu.

Semakin besar Information Gain, berarti fitur itu **semakin bagus** buat misahin data.

---

## ðŸŽ¯ Contoh Sederhana

Misal kamu punya data hewan dan kamu ingin klasifikasi apakah hewan itu **kucing, anjing, atau burung**.

Kamu punya fitur seperti:
- Punya bulu?
- Bisa terbang?
- Suka menggonggong?

Kalau kamu bagi data berdasarkan "bisa terbang", dan semua yang bisa terbang adalah burung, maka kamu langsung tahu jawabannya. Artinya:
- Entropi turun drastis
- Information Gain tinggi banget!
- Fitur itu cocok jadi pertanyaan pertama (akar pohon)

---

## Kenapa Penting?

Dengan Information Gain:
- Pohon keputusan bisa memilih fitur terbaik secara otomatis.
- Hasil klasifikasi jadi lebih akurat dan efisien.
- Model mudah dipahami manusia (karena bentuknya kayak alur logika).

---

## âœ¨ Kesimpulan

- **Decision Tree** itu model ML yang mikirnya kayak kita: tanya satu hal dulu, baru lanjut.
- **Information Gain** bantu milih pertanyaan terbaik buat ditanya di awal.
- Kombinasi keduanya bikin sistem klasifikasi yang **kuat, sederhana, dan gampang dijelaskan**.

### 2. Rumus Entropy dan Information Gain

### Entropy
Rumus:
$$
Entropy(S) = -\sum_{i=1}^{c} p_i \log_2(p_i)
$$
di mana \( p_i \) adalah proporsi kelas ke-i di himpunan S.

### Information Gain
$$
IG(S, A) = Entropy(S) - \sum_{v \in Values(A)} \frac{|S_v|}{|S|} \cdot Entropy(S_v)
$$
- \( A \): atribut (misalnya, X1)
- \( S_v \): subset data dengan nilai atribut \( v \)

---

### 3. Hitung Entropy(S)

Distribusi kelas di seluruh data:
- A: 6
- B: 5
- C: 4

$$
Entropy(S) = -\left(\frac{6}{15} \log_2 \frac{6}{15} + \frac{5}{15} \log_2 \frac{5}{15} + \frac{4}{15} \log_2 \frac{4}{15}\right)
$$

$$
Entropy(S) = -(0.4 \log_2 0.4 + 0.333 \log_2 0.333 + 0.267 \log_2 0.267)
$$

$$
= -(0.4 \times -1.322 + 0.333 \times -1.585 + 0.267 \times -1.90)
$$

$$
= -( -0.5288 - 0.5283 - 0.5073 ) = 1.5644
$$

Jadi,
$$
Entropy(S) \approx 1.56
$$

---

### 4. Hitung Entropy untuk Setiap Nilai X1

#### a. X1 = Tinggi
Data ke: 1, 2, 6, 9, 12, 15 â†’ Total: 6
Kelas: A=6, B=0, C=0

$$
Entropy(Tinggi) = - (1 \cdot \log_2 1) = 0
$$

#### b. X1 = Sedang
Data ke: 3, 5, 8, 11, 14 â†’ Total: 5
Kelas: B=5

$$
Entropy(Sedang) = - (1 \cdot \log_2 1) = 0
$$

#### c. X1 = Rendah
Data ke: 4, 7, 10, 13 â†’ Total: 4
Kelas: C=4

$$
Entropy(Rendah) = - (1 \cdot \log_2 1) = 0
$$

---

### 5. Hitung Information Gain

$$
IG(S, X1) = Entropy(S) - \sum_{v \in \{Tinggi, Sedang, Rendah\}} \frac{|S_v|}{|S|} \cdot Entropy(S_v)
$$

$$
IG(S, X1) = 1.56 - \left( \frac{6}{15} \cdot 0 + \frac{5}{15} \cdot 0 + \frac{4}{15} \cdot 0 \right)
$$

$$
IG(S, X1) = 1.56 - 0 = 1.56
$$

---

### âœ… Kesimpulan

- Entropy(S): **1.56**
- Information Gain untuk fitur **X1**: **1.56**
- Artinya, fitur **X1** dapat memisahkan data secara **sempurna** berdasarkan kelas. Maka X1 adalah fitur yang sangat informatif untuk klasifikasi dalam dataset ini.



```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Data
data = {
    'X1': ['Tinggi', 'Tinggi', 'Sedang', 'Rendah', 'Sedang',
           'Tinggi', 'Rendah', 'Sedang', 'Tinggi', 'Rendah',
           'Sedang', 'Tinggi', 'Rendah', 'Sedang', 'Tinggi'],
    'X2': ['Merah', 'Biru', 'Merah', 'Merah', 'Biru',
           'Hijau', 'Biru', 'Merah', 'Merah', 'Hijau',
           'Hijau', 'Merah', 'Merah', 'Biru', 'Hijau'],
    'Kelas': ['A', 'A', 'B', 'C', 'B',
              'A', 'C', 'B', 'A', 'C',
              'B', 'A', 'C', 'B', 'A']
}

df = pd.DataFrame(data)

# Encode categorical data
le_X1 = LabelEncoder()
le_X2 = LabelEncoder()
le_y = LabelEncoder()

df['X1_enc'] = le_X1.fit_transform(df['X1'])  # 'Rendah'=0, 'Sedang'=1, 'Tinggi'=2
df['X2_enc'] = le_X2.fit_transform(df['X2'])
df['Kelas_enc'] = le_y.fit_transform(df['Kelas'])  # 'A'=0, 'B'=1, 'C'=2

# Train decision tree using only X1 and X2
X = df[['X1_enc', 'X2_enc']]
y = df['Kelas_enc']

clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(X, y)

# Plot the tree
plt.figure(figsize=(10,6))
plot_tree(clf,
          feature_names=['X1', 'X2'],
          class_names=le_y.classes_,
          filled=True,
          rounded=True,
          fontsize=12)
plt.title("Decision Tree menggunakan Entropy (Information Gain)")
plt.show()

# Print decision rules (manually derived)
print("Decision Rules (dari pohon):")
print("1. Jika X1 = 'Tinggi' â†’ Kelas = A")
print("2. Jika X1 = 'Sedang' â†’ Kelas = B")
print("3. Jika X1 = 'Rendah' â†’ Kelas = C")
```

```python
# Data uji baru
X_new = pd.DataFrame({
    'X1': ['Sedang', 'Rendah'],
    'X2': ['Merah', 'Biru']
})

# Encode fitur sesuai encoder yang dilatih sebelumnya
X_new['X1_enc'] = le_X1.transform(X_new['X1'])
X_new['X2_enc'] = le_X2.transform(X_new['X2'])

# Prediksi
X_test_encoded = X_new[['X1_enc', 'X2_enc']]
y_pred = clf.predict(X_test_encoded)

# Tampilkan hasil
for i, pred in enumerate(y_pred):
    print(f"Data {i+1}: X1 = {X_new.loc[i, 'X1']}, X2 = {X_new.loc[i, 'X2']} â†’ Prediksi: {le_y.inverse_transform([pred])[0]}")
```
